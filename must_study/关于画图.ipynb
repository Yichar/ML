```python
#coding:utf-8
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import matplotlib as mpl
# 解决中文显示问题
mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False
```

```python
''' figure 创建一个画图窗口 '''
```

```python
# figure的使用
x = np.linspace(-1, 1, 50)
y1 = 2 * x + 1

# figure 1

plt.figure(1,dpi=40)
plt.plot(x, y1)


# figure 2
y2 = x**2
plt.figure(2,dpi=40)
plt.plot(x, y2)


# figure 3，指定figure的编号并指定figure的大小, 指定线的颜色, 宽度和类型
y2 = x**2
plt.figure(num = 5, figsize = (3, 3))
#先画Y1，再画Y2
plt.plot(x, y1)
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')
plt.show()
```

```python

plt.figure(1,dpi=50)   #创建图表1，dpi为设置图表的大小，默认dpi=80
#2,2,1  表示 2*2 1号位
x = np.linspace(0,10,100)
A=plt.subplot(221)        
B=plt.subplot(222)
C=plt.subplot(223)
D=plt.subplot(224)
#选择A子图
plt.sca(A) 
plt.plot(x,np.exp(x))
plt.sca(D) 
plt.plot(x,np.cos(x))
plt.show()       
```

```python
x=np.arange(0.05,3,0.05)
#常函数
y1=[5 for i in x]
plt.plot(x,y1,linewidth=2,label='常数：y=5')

y2=[2*i for i in x]
plt.plot(x,y2,linewidth=2,label='一次函数：y=2x+1')

y3=[1.5*i*i - 3*i + 1 for i in x]
plt.plot(x,y3,linewidth=2,label='二次函数：y=1.5x$^2$-3x+1')

y4=[np.sin(2*i) for i in x ]
plt.plot(x,y4,linewidth=2,label='正弦函数：y=sin(2x)')
# # loc 设置显示的位置，0是自适应
# # ncol 设置显示的列数
# plt.legend(loc = 0, ncol = 2)

plt.legend(loc = 0, ncol = 1)
plt.show()


```

```python
#哈工程计算机院 2019调剂分数汇总
TJ_dataframe=pd.read_excel("./tiaoji.xlsx")


'''直方图'''
'''
plt.hist(x, bins= 10, range= None, normed= False, weights= None, cumulative= False, bottom= None, histtype= 'bar', align= 'mid',
         orientation= 'vertical', rwidth= None, log= False, color= None, label= None, stacked= False)
x:指定要绘制直方图的数据；
bins：指定直方图条形的个数；
range：指定直方图数据的上下界，默认包含绘图数据的最大值和最小值；
normed：是否将直方图的频数转换成频率；
weights：该参数可为每一个数据点设置权重；
cumulative：是否需要计算累计频数或频率；
bottom：可以为直方图的每个条形添加基准线，默认为0；
histtype：指定直方图的类型，默认为bar，除此还有’barstacked’, ‘step’, ‘stepfilled’；
align：设置条形边界值的对其方式，默认为mid，除此还有’left’和’right’；
orientation：设置直方图的摆放方向，默认为垂直方向；
rwidth：设置直方图条形宽度的百分比；
log：是否需要对绘图数据进行log变换；
color：设置直方图的填充色；
label：设置直方图的标签，可通过legend展示其图例；
stacked：当有多个数据时，是否需要将直方图呈堆叠摆放，默认水平摆放；
'''
plt.figure(1,dpi=100)

#"最低分:"
print("最低分:",TJ_dataframe["score"].min())

print("最高分:",TJ_dataframe["score"].max())
print("中位数分:",TJ_dataframe["score"].median())
print("平均分:",TJ_dataframe["score"].mean())

plt.hist(TJ_dataframe["score"],label="19哈工程计算机调剂分数")  #只要传入数据，直方图就会统计数据出现的次数
plt.legend()
plt.show()
```

```python
'''柱状图'''
#详细 https://blog.csdn.net/qq_29721419/article/details/71638912
#dict 将TJ_dataframe[TJ_dataframe["score"] < 350 强制转换 然后取出score键值
CS_dataframe=pd.read_excel("./cs_first.xlsx")
OT_dataframe=pd.read_excel("./other_first.xlsx")

A=dict(CS_dataframe[CS_dataframe["score"] < 300].count())["score"]
B=dict(CS_dataframe[ ( CS_dataframe["score"] < 310 ) & (CS_dataframe["score"] >= 300) ].count())["score"]
C=dict(CS_dataframe[ (CS_dataframe["score"] < 320) & (CS_dataframe["score"] >= 310) ].count())["score"]
D=dict(CS_dataframe[ (CS_dataframe["score"] < 330) & (CS_dataframe["score"] >= 320) ].count())["score"]
E=dict(CS_dataframe[ (CS_dataframe["score"] < 340) & (CS_dataframe["score"] >= 330) ].count())["score"]
F=dict(CS_dataframe[ (CS_dataframe["score"] < 350) & (CS_dataframe["score"] >= 340) ].count())["score"]
G=dict(CS_dataframe[ (CS_dataframe["score"] < 360) & (CS_dataframe["score"] >= 350) ].count())["score"]
H=dict(CS_dataframe[ (CS_dataframe["score"] >= 360) ].count())["score"]

a=dict(OT_dataframe[OT_dataframe["score"] < 300].count())["score"]
b=dict(OT_dataframe[ (OT_dataframe["score"] < 310) & (OT_dataframe["score"] >= 300) ].count())["score"]
c=dict(OT_dataframe[ (OT_dataframe["score"] < 320) & (OT_dataframe["score"] >= 310) ].count())["score"]
d=dict(OT_dataframe[ (OT_dataframe["score"] < 330) & (OT_dataframe["score"] >= 320) ].count())["score"]
e=dict(OT_dataframe[ (OT_dataframe["score"] < 340) & (OT_dataframe["score"] >= 330) ].count())["score"]
f=dict(OT_dataframe[ (OT_dataframe["score"] < 350) & (OT_dataframe["score"] >= 340) ].count())["score"]
g=dict(OT_dataframe[ (OT_dataframe["score"] < 360) & (OT_dataframe["score"] >= 350) ].count())["score"]
h=dict(OT_dataframe[ (OT_dataframe["score"] >= 360) ].count())["score"]

name_list=["<300","300~310","310~320","320~330","330~340","340~350","350~360",">360"]
CS_class_list=[A,B,C,D,E,F,G,H]
OT_class_list=[a,b,c,d,e,f,g,h]
print("CS最低分:",CS_dataframe["score"].min());print("CS最高分:",CS_dataframe["score"].max())
print("CS中位数分:",CS_dataframe["score"].median());print("CS平均分:",int(CS_dataframe["score"].mean()))
print("其它最低分:",OT_dataframe["score"].min());print("其它最高分:",OT_dataframe["score"].max())
print("其它中位数分:",OT_dataframe["score"].median());print("其它平均分:",int(OT_dataframe["score"].mean()))
plt.figure(dpi=100)

CS_range=[2*i for i in range(len(OT_class_list))]

OT_range=[ i+1 for i in CS_range]    
plt.bar(OT_range,OT_class_list,width=0.8,label="19哈工程计算机其它一志愿分数",fc='b')
plt.bar(CS_range,CS_class_list,width=0.8,tick_label=name_list,label="19哈工程CS学硕一志愿分数",fc='pink')
plt.legend()
plt.show()
```
